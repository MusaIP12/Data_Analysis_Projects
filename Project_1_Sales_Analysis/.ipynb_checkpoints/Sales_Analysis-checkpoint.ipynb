```python
# Data Manipulation
import pandas as pd

# Data Visualisation
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
#Lets check if there are duplicated rows or columns

# Duplicate columns checking
duplicated_cols = sales_df.columns[sales_df.columns.duplicated()]

if len(duplicated_cols) > 0:
    print("Duplicate columns found:")
    print(duplicated_cols)
else:
    print("No duplicate columns found.")
    
# Checking for duplicate Rows
duplicate_rows = sales_df[sales_df.duplicated()]

# If there are any duplicates, 'duplicate_rows' will contain those rows
if not duplicate_rows.empty:
    print("Duplicate rows found:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")
```

```python
# first check if there are columns with missing values

cols_with_missing = [col for col in sales_df.columns
                     if sales_df[col].isnull().any()]

if cols_with_missing:
    print(cols_with_missing)
else:
    print("No columns with misiing values,therefore no need for imputation")

# Find missing values in rows
rows_with_missing = sales_df[sales_df.isnull().any(axis=1)]

if not rows_with_missing.empty:
    print("Rows with missing values found:")
    print(rows_with_missing)
else:
    print("No rows with missing values.")
```

### EXPLORATORY DATA ANALYSIS

```python
# First question is--- what is the overall sales trend
# Lets check the first date of sales and last or max date of the sales
sales_df["order_date"].min()
```

```python
sales_df["order_date"].max()
```

```python
# Getting month year from order_date
sales_df['month_year'] = sales_df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# grouping month_year by sales
sales_df_temp = sales_df.groupby('month_year').sum()['sales'].reset_index()
```

```python
# Setting the figure size
plt.figure(figsize=(16, 5))
plt.plot(sales_df_temp['month_year'], sales_df_temp['sales'], color='#b80045')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

```python
# which are the top 10 products by number of sales
# Grouping products by sales
prod_sales = pd.DataFrame(sales_df.groupby('product_name').sum()['sales'])

# Sorting the dataframe in descending order
prod_sales.sort_values(by=['sales'], inplace=True, ascending=False)

# Top 10 products by sales
prod_sales[:10]
```

```python
# What are the most selling products

# Grouping products by Quantity
best_selling_prods = pd.DataFrame(sales_df.groupby('product_name').sum()['quantity'])

# Sorting the dataframe in descending order
best_selling_prods.sort_values(by=['quantity'], inplace=True, ascending=False)

# Most selling products
best_selling_prods[:10]
```

```python
#Most preffered mode of shipping
# Setting the figure size
plt.figure(figsize=(10, 8))

# countplot: Show the counts of observations in each categorical bin using bars
sns.countplot(x='ship_mode', data=sales_df)

# Display the figure
plt.show()
```

```python
# Whicch are the most profitable products category and sub category

# Grouping products by Category and Sub-Category
cat_subcat = pd.DataFrame(sales_df.groupby(['category', 'sub_category']).sum()['profit'])

# Sorting the values
cat_subcat.sort_values(['category','profit'], ascending=False)
```

```python
# Group data by customer_id and count the number of orders
customer_orders = sales_df.groupby('customer_name')['order_date'].count().reset_index()

# Plot a histogram to visualize the distribution of order frequencies
plt.figure(figsize=(10, 6))
plt.hist(customer_orders['order_date'], bins=20, color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Distribution of Customer Order Frequencies')
plt.xlabel('Number of Orders')
plt.ylabel('Number of Customers')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#import matplotlib.pyplot as plt

# Group data by customer_name and count the total number of orders
customer_orders = sales_df.groupby('customer_name').size().reset_index(name='total_orders')

# Sort customers by total orders in descending order
customer_orders = customer_orders.sort_values(by='total_orders', ascending=False)

# Plot a bar chart to visualize the top customers by total orders
plt.figure(figsize=(12, 6))
plt.bar(customer_orders['customer_name'][:10], customer_orders['total_orders'][:10], color='skyblue')

# Add labels and title
plt.title('Top 10 Customers by Total Orders')
plt.xlabel('Customer Name')
plt.ylabel('Total Orders')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python


# Group data by region and calculate total profit
region_profit = sales_df.groupby('region')['profit'].sum().reset_index()

# Sort regions by total profit in descending order
region_profit = region_profit.sort_values(by='profit', ascending=False)

# Plot a bar chart to visualize the regions contributing more to profit
plt.figure(figsize=(10, 6))
plt.bar(region_profit['region'], region_profit['profit'], color='skyblue')

# Add labels and title
plt.title('Profit Contribution by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Get the top ten regions contributing to profit
top_ten_regions = set(region_profit['region'][:10])

# Get the top ten customers contributing to profit
top_ten_customers = set(customer_orders['customer_name'][:10])

# Check for overlap between top regions and top customers
overlap = top_ten_regions.intersection(top_ten_customers)

# Print the overlap
print("Overlap between top ten regions and top ten customers contributing to profit:", overlap)

```
