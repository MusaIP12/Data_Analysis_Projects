# Checking and handling missing values

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# If you want to drop rows with missing values
#df_cleaned = df.dropna()

# If you want to fill missing values
#df_filled = df.fillna(method='ffill')  # Forward fill missing values
```

```python
df_cleaned = df.drop(columns=['Original Res Forecast before Lockdown'])
```

```python
df_cleaned.head()
```

```python
# Lets convert datte column to datetime format
df_cleaned['Date Time Hour Beginning'] = pd.to_datetime(df_cleaned['Date Time Hour Beginning'])

# Create separate Date and Time columns
df_cleaned['Date_Time'] = df_cleaned['Date Time Hour Beginning'].dt.date
df_cleaned['Time_Hr'] = df_cleaned['Date Time Hour Beginning'].dt.strftime('%H:%M:%S')

# Optional: Drop the original column if no longer needed
df_cleaned.drop(columns=['Date Time Hour Beginning'], inplace=True)
```

```python
df_cleaned.info()
```

```python
df_cleaned = df_cleaned[['Date_Time', 'Time_Hr'] + [col for col in df_cleaned.columns if col not in ['Date_Time', 'Time_Hr']]]
```

```python

df_cleaned['Date_Time'] = pd.to_datetime(df_cleaned['Date_Time'])
```

```python
df_cleaned.head()
```

```python
df_cleaned['Time_Hr'] = pd.to_datetime(df_cleaned['Time_Hr'], format='%H:%M:%S').dt.time
```

```python
df_cleaned.info()
```

## SEGMENT THE DATA (DURING and POST COVID)

```python
# Lets divide the datasets

# During COVID: data from April 1, 2020, to June 22, 2022
during_covid = df_cleaned[(df_cleaned['Date_Time'] >= '2020-04-01') & (df_cleaned['Date_Time'] <= '2022-06-22')]

# Post-COVID: data from June 23, 2022, onwards
post_covid = df_cleaned[df_cleaned['Date_Time'] >= '2022-06-23']
```
